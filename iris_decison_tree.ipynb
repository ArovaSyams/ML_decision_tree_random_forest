```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

### Dataset

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
np.shape(X)
```

### creating decision tree

```python
# criterion have "gini", "entropy"
# ccp_alpha is density for the pruning
clf = DecisionTreeClassifier(criterion="entropy", ccp_alpha=0)

clf = clf.fit(X,y)

clf.predict([[6.1, 4.4, 1.5, 5.2]])
```

### Visuallize

```python
plot_tree(clf)
```

```python
plt.figure(figsize=(15, 12))
plot_tree(clf, filled=True, feature_names=["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"],
    class_names=["Iris-setosa", "Iris-versicolor", "Iris-Virginica"])
plt.show()
```
