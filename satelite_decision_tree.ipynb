```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
train_data = pd.read_csv("sat_train.csv")
```

```python
train_data.head()
```

```python
inputs = train_data.drop(["Column37"], axis=1)
targets = train_data["Column37"]
```

```python
targets
```

```python
clf = DecisionTreeClassifier(random_state=365, ccp_alpha=0.02)
```

```python
clf.fit(inputs, targets)
```

```python
plot_tree(clf)
```

```python
plt.figure(figsize=(15, 12))
plot_tree(clf, filled=True)
plt.show()
```
